```python
# import libraries
import numpy as np
import plotly.express as px  
import skimage as ski
from skimage.color import rgb2hsv, hsv2rgb

# take the image in RGB
rgb_img = ski.data.cat()

img = px.imshow(rgb_img)
img.show()

# convert the image in HSV
hsv_img = rgb2hsv(rgb_img)

# Function that converts a hsv image in a gray_scale hsv image
def gray_scale(hsv):
    hsv[:, :, 1] = 0
    return hsv

# get the image's gray scale
grayScale_img = gray_scale(hsv_img)
img = px.imshow(hsv2rgb(grayScale_img))
img.show()

# set the bounds for the selected hue
lower_bound = np.array([0,0,100]) # should be white
upper_bound = np.array([0,40,100])
```
